# Free Mobile Apps Analysis

 1. In this project, I will analyze data from apps on Google Play and the App Store. The company builds free apps, and its main source of revenue comes from in-app ads.

 2. The goal of this project is to identify what types of apps attract more users. This project also focuses on practicing Python fundamentals by avoiding libraries such as Pandas and working mainly with core Python features.

```python
from csv import reader
```

```python
def extract_data(data_set):
    data = open(data_set,encoding='utf-8')
    raw = reader(data)
    table = list(raw)
    data.close()
    return table
  
    
```

```python
def explore_data(data_set,start,end,columns=False):
    data_used = data_set[start:end]
    for x in data_used:
        print(x)
        print('\n')

    if columns:
        print(f'rows {len(data_set)}')
        print(f'columns {len(data_set[0])}')
```

```python
df1 = extract_data('AppleStore.csv')
explorer = explore_data(df1,0,10)
print(explorer)
```
